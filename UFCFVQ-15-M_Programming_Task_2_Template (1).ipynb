### Requirement FR9 - Read CSV data from two files and merge it into a single Data Frame 

```python
# import modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# add code here
# pd.set_option('max_rows', None)
```

```python
# reading csv files into the dataframe
df1 = pd.read_csv('task2a (1).csv')
df2 = pd.read_csv('task2b (1).csv')
```

```python
df1.head()  # Displays first five rows
```

```python
df = df1.merge(df2, 'left', 'region')
df.head()
```
